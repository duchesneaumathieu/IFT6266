##Some maths

Let,
\begin{align*}
    h^a &= W^{(1)} x + b^{(1)}\\
    h^s &= sigmoid(h^a)\\
    o^a &= W^{(2)} h^s + b^{(2)}\\
    o^s &= softmax(o^a)\\
\end{align*}
with,
$$d = dim(x),\ d_h = dim(h^a) = dim(h^s),\ d_m = dim(o^a) = dim(o^s)$$

###Softmax gradient

Using cross entropy we have,
\begin{align*}
    L(x, y) \ &= -\log(o^s_y)\\
    &= - \log \left( softmax(o^a_y) \right)\\
    &= - \log \left( \frac{e^{o^a_y}}{\sum_{t} e^{o^a_t}} \right) \\
    &= \log \left( \sum_{t} e^{o^a_t} \right) - o^a_y  \\
\end{align*}
Then,
\begin{eqnarray*}
    \frac{\partial L(x, y)}{\partial o^a_{y'}} &=& \frac{\partial \left(\log \left( \sum_{t} e^{o^a_t} \right) - o^a_y\right) }{\partial o^a_{y'}}\\
    &=& \frac{\frac{\partial}{\partial o^a_{y'}} \sum_{t} e^{o^a_t}}{\sum_{t} e^{o^a_t}} - \frac{\partial o^a_y }{\partial o^a_{y'}}\\
    &=& \frac{e^{o^a_{y'}}}{\sum_{t} e^{o^a_t}} - I_{y=y'}\\
    &=& softmax(o^a_{y'}) - I_{y=y'}\\
    &=& o^s_{y'} - I_{y=y'}\\
\end{eqnarray*}
So,
$$ \frac{\partial L}{\partial o^a} = o^s - onehot_m(y) $$

###Second layer gradient
\begin{eqnarray*}
    \frac{\partial L}{\partial W^{(2)}_{k,j}} &=& \frac{\partial L}{\partial o^a_k} \frac{\partial o^a_k}{\partial W^{(2)}_{k,j}}\\
    &=&\frac{\partial L}{\partial o^a_k} \frac{\partial}{\partial W^{(2)}_{k,j}} \sum_{j'} W^{(2)}_{k,j'}h^s_{j'} + b^{(2)}_k\\
    &=&\frac{\partial L}{\partial o^a_k} h^s_{j}
\end{eqnarray*}
and,
\begin{eqnarray*}
    \frac{\partial L}{\partial b^{(2)}_k} &=& \frac{\partial L}{\partial o^a_k} \frac{\partial o^a_k}{\partial b^{(2)}_k}\\
    &=&\frac{\partial L}{\partial o^a_k} \frac{\partial}{\partial b^{(2)}_k} \sum_{j'} W^{(2)}_{k,j'}h^s_{j'} + b^{(2)}_k\\
    &=&\frac{\partial L}{\partial o^a_k}
\end{eqnarray*}
so,
$$\frac{\partial L}{\partial W^{(2)}} = \frac{\partial L}{\partial o^a} (h^s)^T$$
and,
$$\frac{\partial L}{\partial b^{(2)}} = \frac{\partial L}{\partial o^a}$$

###Sigmoid gradient

\begin{align*}
    \frac{\partial L}{\partial h^s_j} &= \sum^{d_m}_{k=1} \frac{\partial L}{\partial o^a_k} \frac{\partial o^a_k}{\partial h^s_j}\\
    &= \sum^{d_m}_{k=1} \frac{\partial L}{\partial o^a_k} \frac{\partial}{\partial h^s_j} \sum^{d_h}_{j'=1} W^{(2)}_{k,j'} h^s_{j'} + b^{(2)}_k\\
    &= \sum^{d_m}_{k=1} \frac{\partial L}{\partial o^a_k}  W^{(2)}_{k,j}\\
\end{align*}

Also,
\begin{align*}
    \frac{\partial L}{\partial h^a_i} &= \frac{\partial L}{\partial h^s_i} \frac{\partial h^s_i}{\partial h^a_i}\\
    &= \frac{\partial L}{\partial h^s_i} \frac{\partial}{\partial h^a_i} \frac{1}{1+e^{-h^a_i}}\\
    &= \frac{\partial L}{\partial h^s_i} \frac{e^{-h^a_i}}{(1+e^{-h^a_i})^2}\\
    &= \frac{\partial L}{\partial h^s_i}  \frac{sigmoid(h^a_i)^2}{e^{h^a_i}}\\
    &= \frac{\partial L}{\partial h^s_i}  \frac{(h^s_i)^2}{e^{h^a_i}}\\
\end{align*}
So,
$$\frac{\partial L}{\partial h^s} = (W^{(2)})^T \frac{\partial L}{\partial o^a}$$
and
$$\frac{\partial L}{\partial h^a} = \frac{\partial L}{\partial h^s} \odot \frac{(h^s)^2}{e^{h^a}}$$
with $\odot$ the element wise multiplication.

###First layer gradient
\begin{eqnarray*}
    \frac{\partial L}{\partial W^{(1)}_{k,j}} &=& \frac{\partial L}{\partial h^a_k} \frac{\partial h^a_k}{\partial W^{(1)}_{k,j}}\\
    &=&\frac{\partial L}{\partial h^a_k} \frac{\partial}{\partial W^{(1)}_{k,j}} \sum_{j'} W^{(1)}_{k,j'}x_{j'} + b^{(1)}_k\\
    &=&\frac{\partial L}{\partial h^a_k} x_{j}
\end{eqnarray*}
and,
\begin{eqnarray*}
    \frac{\partial L}{\partial b^{(1)}_k} &=& \frac{\partial L}{\partial h^a_k} \frac{\partial h^a_k}{\partial b^{(1)}_k}\\
    &=&\frac{\partial L}{\partial h^a_k} \frac{\partial}{\partial b^{(1)}_k} \sum_{j'} W^{(1)}_{k,j'}x_{j'} + b^{(1)}_k\\
    &=&\frac{\partial L}{\partial h^a_k}
\end{eqnarray*}
so,
$$\frac{\partial L}{\partial W^{(1)}} = \frac{\partial L}{\partial h^a} x^T$$
and,
$$\frac{\partial L}{\partial b^{(1)}} = \frac{\partial L}{\partial h^a}$$

##Starting

```python
import gzip
import pickle
import numpy as np
data = pickle.load(gzip.open('mnist.pkl.gz'), encoding='latin1')
```

###Loading MNIST

```python
train_x = data[0][0] #: matrice de train data
train_y = data[0][1] #: vecteur des train labels

valid_x = data[1][0] # : matrice de valid data
valid_y = data[1][1] #: vecteur des valid labels

test_x = data[2][0] #: matrice de test data
test_y = data[2][1] #: vecteur des test labels

train_ohy = np.zeros((train_x.shape[0], 10))

for i, j in enumerate(train_y):
    train_ohy[i, j] = 1

valid_ohy = np.zeros((valid_x.shape[0], 10))

for i, j in enumerate(valid_y):
    valid_ohy[i, j] = 1
        
test_ohy = np.zeros((test_x.shape[0], 10))

for i, j in enumerate(test_y):
    test_ohy[i, j] = 1
```

###MLP

```python
#for numerically stable softmax
#https://timvieira.github.io/blog/post/2014/02/11/exp-normalize-trick/

def softmax(x):
    v = np.exp(x - x.max(axis=1, keepdims=True))
    return  v / v.sum(axis=1, keepdims=True)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def accuracy(f, x, y):
    return (f(x)==y).mean()
   

class MLP:
    def __init__(self, d, dh, dm):
        self.d = d
        self.dh = dh
        self.dm = dm
        c1 = 0.1 * np.sqrt(1 / (d + dh))
        c2 = 0.1 * np.sqrt(1 / (dh + dm))
        self.w1 = np.random.uniform(-c1, c1, (d, dh))
        self.b1 = np.zeros((1, dh))
        self.w2 = np.random.uniform(-c2, c2, (dh, dm))
        self.b2 = np.zeros((1, dm))
        self.n_epoch = 0
        self.memory = [0 for i in range(10)]
        
    def f_prop(self, x):
        ha = np.dot(x, self.w1) + self.b1
        hs = sigmoid(ha)
        oa = np.dot(hs, self.w2) + self.b2
        os = softmax(oa)
        return [os,oa,hs,ha]
    
    def b_prop(self, x, y):
        os,oa,hs,ha = self.f_prop(x)
        goa = os - y
        gw2 = np.dot(hs.transpose(), goa)/x.shape[0]
        gb2 = goa.mean()
        ghs = np.dot(goa, self.w2.transpose())
        gha = ghs*hs*hs/np.exp(ha)
        gw1 = np.dot(x.transpose(), gha)/x.shape[0]
        gb1 = gha.mean()
        return [gw1,gb1,gw2,gb2]
    
    def _next_batch(self, x, y, batch_size, i):
        start = (batch_size * i) % x.shape[0]
        end = min(start + batch_size, x.shape[0])
        return x[start:end, :], y[start:end, :]
    
    def _gradient_step(self, x, y, alpha):
        gw1,gb1,gw2,gb2 = self.b_prop(x,y)
        self.w1 -= alpha*gw1
        self.b1 -= alpha*gb1
        self.w2 -= alpha*gw2
        self.b2 -= alpha*gb2

    def _gradient_epoch(self, x, y, batch_size, alpha):
        self.n_epoch += 1
        i=0
        while(i*batch_size < x.shape[0]):
            batch_x, batch_y = self._next_batch(x, y, batch_size, i)
            self._gradient_step(batch_x, batch_y, alpha)
            i += 1
            
    def _verbose_init(self):
        print("\t\t Train cost\t\t Train acc\t\t Test cost\t\t Test acc")
        
    def print_stats(self, X, Y, TX, TY):
        print(self.n_epoch, "\t\t", "%.8f" % self.cost(X,Y),
              "\t\t", "%.7f" % accuracy(self.predict, X, np.argmax(Y,axis=1)),
              "\t\t", "%.7f" % self.cost(TX,TY),
              "\t\t", "%.6f" % accuracy(self.predict, TX, np.argmax(TY,axis=1)))
    
    def set_memory(size):
        if n_epoch == 0: self.memory = [0 for i in range(size)]
            
    def gradient_descent(self, X, Y, VX, VY, alpha, batch_size, max_epoch=100, tol=-1e-3, verbose=False, test_x=None, test_ohy=None):
        if verbose: self._verbose_init()
        if verbose: self.print_stats(X,Y,test_x,test_ohy)
        if self.n_epoch == 0: self.last_cost = self.cost(VX, VY)
        while(self.n_epoch < len(self.memory)):
            self._gradient_epoch(X, Y, batch_size, alpha)
            if verbose: self.print_stats(X,Y,test_x,test_ohy)
            cost = self.cost(VX, VY)
            self.memory[self.n_epoch-1] = cost - self.last_cost
            self.last_cost = cost
        while(self.n_epoch < max_epoch):
            if np.mean(self.memory) > tol: break
            self._gradient_epoch(X, Y, batch_size, alpha)
            if verbose: self.print_stats(X,Y,test_x,test_ohy)
            cost = self.cost(VX, VY)
            self.memory[(self.n_epoch-1)%len(self.memory)] = cost - self.last_cost
            self.last_cost = cost
    
    def predict_proba(self, x):
        return self.f_prop(x)[0]
    
    def predict(self, x):
        return np.argmax(self.predict_proba(x), axis=1)
    
    def cost(self, x, y):
        return -np.log(np.sum(self.predict_proba(x)*y, axis=1)).mean()
    
    def num_grad(self, x, y, e=1e-10):
        theta = [self.w1, self.b1, self.w2, self.b2]
        grad = [np.zeros(self.w1.shape),np.zeros(self.b1.shape),np.zeros(self.w2.shape),np.zeros(self.b2.shape)]
        for k in range(len(theta)):
            for i in range(theta[k].shape[0]):
                for j in range(theta[k].shape[1]):
                    theta[k][i][j] -= e
                    c1 = self.cost(x, y)
                    theta[k][i][j] += 2*e
                    c2 = self.cost(x, y)
                    grad[k][i][j] = (c2-c1)/(2*e)
                    theta[k][i][j] -= e
        return grad
    
    def close_enough_grad(self, x, y, tol=1e-5):
        grad = self.b_prop(x, y)
        ngrad = self.num_grad(x, y)
        ans = True
        for i in range(4): ans = ans and (np.abs(grad[0]-ngrad[0]) < tol).all()
        return ans
```

###Gradient check

```python
mlp = MLP(784,100,10)
mlp.close_enough_grad(train_x[[0],:],train_ohy[[0],:])
```

```python
mlp.gradient_descent(train_x, train_ohy, valid_x, valid_ohy,0.4,1000,max_epoch=250,tol=1e-3,verbose=True,test_x=test_x,test_ohy=test_ohy)
```

###Using relu

```python
class MLPr(MLP):
    def f_prop(self, x):
        ha = np.dot(x, self.w1) + self.b1
        hs = np.maximum(0, ha) #new line
        oa = np.dot(hs, self.w2) + self.b2
        os = softmax(oa)
        return [os,oa,hs,ha]
    
    def b_prop(self, x, y):
        os,oa,hs,ha = self.f_prop(x)
        goa = os - y
        gw2 = np.dot(hs.transpose(), goa)/x.shape[0]
        gb2 = goa.mean()
        ghs = np.dot(goa, self.w2.transpose())
        gha = np.multiply(ghs, (hs > 0)) #new line
        gw1 = np.dot(x.transpose(), gha)/x.shape[0]
        gb1 = gha.mean()
        return [gw1,gb1,gw2,gb2]   
```

```python
mlpr = MLPr(784,100,10)
mlpr.close_enough_grad(train_x[[0],:],train_ohy[[0],:])
```

```python
mlpr.gradient_descent(train_x, train_ohy, valid_x, valid_ohy,0.4,1000,max_epoch=250,tol=1e-3,verbose=True,test_x=test_x,test_ohy=test_ohy)
```

###Pre-trainning and more capacity

```python
from sklearn.decomposition import PCA
```

```python
mu = np.mean(train_x)
sig = np.std(train_x)
sig += (sig==0)
pca = PCA(n_components=300)
pca.fit((train_x-mu)/sig)
def pretrain(x):
    return pca.transform((x-mu)/sig)
pntrain_x = pretrain(train_x)
pnvalid_x = pretrain(valid_x)
pntest_x = pretrain(test_x)
```

```python
mlprpn = MLPr(300, 400, 10)
```

```python
mlprpn.gradient_descent(pntrain_x, train_ohy, pnvalid_x, valid_ohy,0.4,1000,max_epoch=250,tol=1e-3,verbose=True,test_x=pntest_x,test_ohy=test_ohy)
```
